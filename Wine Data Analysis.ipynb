# GATHERING DATA

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
red_wine = pd.read_csv('winequality-red.csv', sep = ';')
white_wine = pd.read_csv('winequality-white.csv', sep = ';')
```

```python
# CREATING ARRAY FOR INSERTION IN THE COLUMN

color_red = np.repeat('RED', red_wine.shape[0])
color_white = np.repeat('WHITE',white_wine.shape[0])
```

```python
# CREATING COLUMN AND INSERTING THE VALUES IN THE RED_WINE DATASET
red_wine['color'] = color_red
```

```python
# CREATING COLUMN AND INSERTING THE VALUES IN THE RED_WINE DATASET
white_wine['color'] = color_white
```

```python
# APPENDING VALUE OF WHITE WINE DATASET INTO RED_WINE DATASET
combine_wine = red_wine.append(white_wine)
combine_wine
```

```python
# SAVING THE COMBINED DATASET
combine_wine.to_csv('wine_data_combined.csv', index = False)
```

# DRAWING CONCLUSIONS USING GROUPBY

### Is a certain type of wine associated with higher quality?


```python
combine_wine.groupby('color')['quality'].mean()
```

### What level of acidity receives the highest average rating?

```python
combine_wine.describe().pH
```

```python
# creating range using cut function of pandas !!
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]
bin_names = ['High', 'Mod_high', 'Medium', 'Low']

combine_wine['Acidity Level'] = pd.cut(combine_wine['pH'], bin_edges, labels = bin_names)
combine_wine.head()
```

```python
combine_wine.groupby('Acidity Level')['pH'].mean()
```

### Selecting the data with pH level greater than 3.51

```python
ph_level = combine_wine.query('pH > 3.50')
ph_level
```

# VISUALIZATION

```python
colors = ['red','white']
color_means = combine_wine.groupby('color')['quality'].mean()
color_means.plot(kind = 'bar', title = 'Average wine quality by color', color = colors, alpha = .7)

plt.xlabel('COLOR', fontsize = 20)
plt.ylabel('QUALITY', fontsize = 20)
```

```python
counts = combine_wine.groupby(['quality', 'color']).count()['pH']
counts.plot(kind = 'bar', title = 'Counts by Wine Color and Quality', color = colors, figsize =(15,15))
plt.xlabel('Quality and Color', fontsize = 20)
plt.ylabel('Counts', fontsize = 20)
```

```python
# get counts for each rating and color
color_counts = combine_wine.groupby(['color', 'quality']).count()['pH']
color_counts
```

```python
# get total counts for each color
color_totals = combine_wine.groupby('color').count()['pH']
color_totals
```

```python
# get proportions by dividing red rating counts by total # of red samples
red_proportions = color_counts['RED'] / color_totals['RED']
red_proportions
```

```python
# get proportions by dividing white rating counts by total # of white samples
white_proportions = color_counts['WHITE'] / color_totals['WHITE']
white_proportions
```

```python
red_proportions['9'] = 0
red_proportions
```

```python
ind = np.arange(len(red_proportions))  # the x locations for the groups
width = 0.35       # the width of the bars
```

```python
# plot bars
red_bars = plt.bar(ind, red_proportions, width, color='r', alpha=.7, label='Red Wine')
white_bars = plt.bar(ind + width, white_proportions, width, color='w', alpha=.7, label='White Wine')

# title and labels
plt.ylabel('Proportion')
plt.xlabel('Quality')
plt.title('Proportion by Wine Color and Quality')
locations = ind + width / 2  # xtick locations
labels = ['3', '4', '5', '6', '7', '8', '9']  # xtick labels
plt.xticks(locations, labels)

# legend
plt.legend()
```
